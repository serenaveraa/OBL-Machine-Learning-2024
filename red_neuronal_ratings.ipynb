```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

# Cargar los datos
data = pd.read_csv('./dataset/train.csv') # Cambia esto por la ruta de tu archivo CSV.


def convert_to_numeric(value):
    if isinstance(value, str):
        if 'K' in value:
            return float(value.replace('K', '')) * 1_000
        elif 'M' in value:
            return float(value.replace('M', '')) * 1_000_000
    return float(value)

# Preprocesar los datos
# Rellenar valores faltantes en las columnas numéricas con la media
# Convertir valores con sufijos 'K' o 'M' a números

def preporcess_data(data):
    columns_with_suffixes = ['Number of Reviews', 'Wishlist']  # Cambia estas columnas si es necesario
    for col in columns_with_suffixes:
        data[col] = data[col].apply(convert_to_numeric)

    # Preprocesar los datos
    # Rellenar valores faltantes en las columnas numéricas con la media
    data.fillna(data.mean(numeric_only=True), inplace=True)

    # Rellenar valores faltantes en las columnas categóricas con una cadena 'Unknown'
    data.fillna('Unknown', inplace=True)

    # Convertir columnas categóricas en valores numéricos
    categorical_columns = ['Game Title', 'Publisher', 'Platform', 'Genre']
    label_encoders = {}

    for col in categorical_columns:
        le = LabelEncoder()
        data[col] = le.fit_transform(data[col])
        label_encoders[col] = le
    return data

data = preporcess_data(data)
# Seleccionar características (X) y objetivo (y)
X = data.drop(['id', 'Rating', 'Summary'], axis=1)  # Excluye ID, Rating (objetivo), y texto.
y = data['Rating']

# Escalar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Construir la red neuronal
model = Sequential([
    Dense(128, activation='relu', input_dim=X_train.shape[1]),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dense(1)  # Salida para predicción de un valor continuo
])

# Compilar el modelo
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Entrenar el modelo
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2, verbose=1)

# Evaluar el modelo
loss, mae = model.evaluate(X_test, y_test, verbose=0)
print(f"Mean Absolute Error on test set: {mae:.2f}")

# Predicción con nuevos datos (opcional)
predictions = model.predict(X_test)





#evaluar test
# Cargar el archivo de prueba
test_data = pd.read_csv('./dataset/test.csv')

# Preprocesar los datos de prueba
# Convertir valores con sufijos 'K' o 'M' a números en las mismas columnas que el entrenamiento
data = preporcess_data(test_data)

# Seleccionar características (X) para el conjunto de prueba
X_test = test_data.drop(['id'], axis=1)  # Excluye columnas innecesarias

# Escalar los datos de prueba con el mismo escalador
X_test_scaled = scaler.transform(X_test)

# Usar el modelo entrenado para predecir
predictions = model.predict(X_test_scaled)

# Añadir las predicciones al dataframe de prueba
test_data['Predicted Rating'] = predictions

output = test_data[['id']]
output['Predictions'] = predictions

# Guardar los resultados en un archivo CSV
output.to_csv('test_predictions.csv', index=False)

print("Predictions saved to 'test_predictions.csv' with id and Predicted Rating only.")
```
